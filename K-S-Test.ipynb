```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#creating random variable which follows gaussian distribution
x=stats.norm.rvs(size=1000)
print(len(x),type(x))
```

```python
sns.set_style("whitegrid")
sns.kdeplot(x[:])
```

```python
stats.kstest(x,'norm')
```

#### p-value is 66% which is significant thus null hypothesis is true and the random variable x follows normal distribution.
