```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#creating random variable x which follows gaussian distribution
x=stats.norm.rvs(size=1000)
print(len(x),type(x))
```

```python
sns.set_style("whitegrid")
sns.kdeplot(x[:])
```

```python
stats.kstest(x,'norm')
```

#### p-value is 66% which is significant thus null hypothesis is true and the random variable x follows normal distribution.

```python
#creating random variable y which follows uniform distribution
y=np.random.uniform(0,1,10000)
print(len(y),type(y))
```

```python
sns.set_style("whitegrid")
sns.kdeplot(y[:])
```

```python
stats.kstest(y,'norm')
```

#### p-value is very small thus null hypothesis is false and the random variable y does not follow normal distribution.
